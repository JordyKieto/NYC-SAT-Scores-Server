```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import pdb

driver = webdriver.Chrome(executable_path='./chromedriver')
url = "https://chalkbeat.org/wp-content/static/shortcodes/searchfield/?sheet=publish&key=1A1tknyJWD20gu1qrBdvrzFY2yEhazOY5H-B_ImgMhSc&initialWidth=650&childId=1A1tknyJWD20gu1qrBdvrzFY2yEhazOY5H-B_ImgMhSc&parentTitle=Find%20out%20what%20your%20New%20York%20City%20school%20spends%20per%20student&parentUrl=https%3A%2F%2Fchalkbeat.org%2Fposts%2Fny%2F2018%2F08%2F31%2Fhow-does-your-schools-budget-compare-to-others-in-new-york-city%2F"
driver.get(url)
school_spend_df = {'School Name': [], 'Spent Per Student': [], 'Student enrollment': [], 'Amount needed to reach 100% Fair Student Funding': [], 'Percent of Fair Student Funding': []} 
```

```python
name_field = driver.find_element_by_css_selector('body > div > div.instructions > div > div.choices__inner > input')
name_field.click()
```

```python
schools = driver.find_elements_by_css_selector("body > div > div.instructions > div > div.choices__list.choices__list--dropdown.is-active > div > div")
schools = [c.text for c in schools]
```

```python
total = len(schools) - 1
for index, school in enumerate(schools):
    name_field.send_keys(school)
    name_field.send_keys(u'\ue007')
    school_tables = driver.find_elements_by_css_selector('#datatables > div')
    for _ in school_tables:
        percent_spent = school_tables[0].find_element_by_css_selector('div > table > tbody > tr:nth-child(2) > td:nth-child(2)').text
        school_name = school_tables[0].find_element_by_css_selector('div > table > thead > tr > th').text
        school_spend_df['School Name'] += [school_name]
        school_spend_df['Spent Per Student'] += [percent_spent]
        school_spend_df['Amount needed to reach 100% Fair Student Funding'] += [school_tables[0].find_element_by_css_selector('div > table > tbody > tr:nth-child(4) > td:nth-child(2)').text]
        school_spend_df['Student enrollment'] += [school_tables[0].find_element_by_css_selector('div > table > tbody > tr:nth-child(1) > td:nth-child(2)').text]
        school_spend_df['Percent of Fair Student Funding'] += [school_tables[0].find_element_by_css_selector('div > table > tbody > tr:nth-child(3) > td:nth-child(2)').text]
        print(index, ' / ', total, school)
    name_field.send_keys(Keys.BACKSPACE)
```

```python
school_spend_df = pd.DataFrame(school_spend_df)
school_spend_df.to_csv('school_spend.csv')
```
